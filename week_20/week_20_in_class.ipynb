Take the result of the following line of code (but use whatever available text you want)
and write a function to concatenate the data back into sentences, with one readable
sentence in each list.


```python
import nltk
```

```python
sentences = nltk.corpus.gutenberg.sents("carroll-alice.txt")[:10]
sentences
```

```python
def untokenize(sentences):
    my_list = []
    for sentence in sentences:
        my_string =''
        for word in sentence:
            my_string += ' ' + word
        placeholder = []
        placeholder.append(my_string)
        my_list.append(placeholder)
    return my_list
    
untokenize(sentences)
```

```python
starter = [1,2,'a', 'b', 7.6]
other = [True, 'a', 46, 982.0, False]

def numeric(a_list):
    output = []
    for i in a_list: 
        if (type(i) != int and type(i) != float): 
            output.append(i)
    return output

numeric(starter)


numeric(other)
```
