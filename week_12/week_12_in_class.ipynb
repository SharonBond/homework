```python
df = pd.read_csv('temperature.csv')
df['datetime'] = pd.to_datetime(df['datetime'])
df.head()
```

```python
df["Chi_is_freezing"] = df.Chicago.map(lambda x: x<=273.15)
df = df.assign(Chi_is_freezing = lambda x: x['Chicago']<=273.15)
df
```
